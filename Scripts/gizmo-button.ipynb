```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('gizmo-bounding.jpg', 0)

plt.imshow(image)
```

```python
gray_filtered = cv2.inRange(image, 230, 255)
g2 = gray_filtered/255 * 2
plt.imshow(g2)
```

```python
gray_filtered = cv2.inRange(image, 190, 220)
g1 = gray_filtered/255 * 1
plt.imshow(g1)
```

```python
gray_filtered = cv2.inRange(image, 160, 180)
g3 = gray_filtered/255 * 3
plt.imshow(g3)
```

```python
gray_filtered = cv2.inRange(image, 130, 150)
g4 = gray_filtered/255 * 4
plt.imshow(g4)
```

```python
gray_filtered = cv2.inRange(image, 90, 115)
g5 = gray_filtered/255 * 5
plt.imshow(g5)
```

```python
gray_filtered = cv2.inRange(image, 60, 85)
g6 = gray_filtered/255 * 6
plt.imshow(g6)
```

```python
g = cv2.bitwise_or(g1,g2)
g = cv2.bitwise_or(g,g3)
g = cv2.bitwise_or(g,g4)
g = cv2.bitwise_or(g,g5)
g = cv2.bitwise_or(g,g6)
```

```python
plt.imshow(g)
```

```python
cv2.imwrite("gizmo-bounding.jpg", g)
```

```python
g = g * 255/6.0
```

```python
plt.imshow(g)
```
